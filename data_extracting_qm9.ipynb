```python
import pandas as pd
import urllib.request
import os
```

```python
url = "https://deepchemdata.s3-us-west-1.amazonaws.com/datasets/qm9.csv"
csv_raw = "qm9_raw.csv"
```

```python
if not os.path.exists(csv_raw):
    urllib.request.urlretrieve(url, csv_raw)
df = pd.read_csv(csv_raw)
print(f"Total molecules: {len(df)}")
print(f"Columns: {list(df.columns)}")
print(df.head())
```

```python
features = {
    "mol_id":   "mol_id",       
    "smiles":   "smiles",      
    "mu":       "dipole_moment",  
    "alpha":    "polarizability", 
    "homo":     "homo_energy",    
    "lumo":     "lumo_energy",   
    "gap":      "homo_lumo_gap",
    "r2":       "spatial_extent",
    "zpve":     "zero_point_energy",    
    "u0":       "internal_energy_0K",    
    "cv":       "heat_capacity",    
    "u298":     "internal_energy_298K",  
    "h298":     "enthalpy_298K",          
    "g298":     "free_energy_298K",
}
df_clean = df[list(features.keys())].rename(columns=features)
print(f"\nShape: {df_clean.shape}")  
print(df_clean.describe())
```

```python
output_path = "molecule_data.csv"
df_clean.to_csv(output_path, index=False)
print(f"Saved {len(df_clean)} molecules × {df_clean.shape[1]} features to '{output_path}'")
```
